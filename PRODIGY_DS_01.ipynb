```python
# importing the necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing the dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# Plotting a count of passengers by gender ('sex')

sns.countplot(x='sex', data=titanic)
plt.show()
```

```python
# Plotting a count of passengers based on their port of embarkation
sns.countplot(x='embarked', data=titanic)
plt.show()
```

```python
# Plotting the distribution of passenger ages, differentiating by gender
sns.histplot(data=titanic, x='age', hue='sex', kde=True)
plt.show()
```

```python
# a histogram with hue for survival
sns.histplot(data=titanic, x='age', hue='survived', multiple='stack', kde=True)
plt.title('Distribution of Age by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# histogram for each gender
g = sns.FacetGrid(titanic, col='sex', hue='survived', height=5)
g.map(sns.histplot, 'age', kde=True)
g.add_legend(title='Survived')
plt.suptitle('Distribution of Age by Survival and Gender', y=1.05)
plt.show()
```

```python
# Creating age groups and visualizing survival counts across these groups by gender

titanic['age_group'] = pd.cut(titanic['age'], bins=[0, 18, 30, 50, 80], labels=['Child', 'Young Adult', 'Adult', 'Senior'])

sns.catplot(data=titanic, x='age_group', hue='sex', col='survived', kind='count')
plt.show()
```
